```python
import pandas as pd
df = pd.read_excel("stock_data.xlsx", "Sheet2")
df
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell

df = pd.read_excel("stock_data.xlsx", "Sheet2", converters = {
    'people':convert_people_cell
})

df
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell=="not available":
        return None
    return cell
df = pd.read_excel("stock_data.xlsx", "Sheet2", converters = {
    'people':convert_people_cell,
    'eps':convert_eps_cell
})

df
```

```python
df.to_excel("new.xlsx", sheet_name="stocks",startrow=1, startcol=2)
```

```python
df.to_excel("new.xlsx", sheet_name="stocks",startrow=1, startcol=2, index=False)
```
